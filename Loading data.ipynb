Loading Data in Python

```python
import numpy as np
import pickle
import pandas as pd

filename = "load.csv"
```

```python
## HEader

cols = None
data = []

with open(filename) as f:
    for line in f.readlines():
        vals = line.replace("\n","").split(",")
        if cols is None:
            cols = vals
        else:
            data.append([float(x) for x in vals])
            
d0 = pd.DataFrame(data,columns=cols)
print(d0.dtypes)
d0.head()
```

```python
#np.loadtxt does not deal well with characters so we skip the first row
d1 = np.loadtxt(filename, skiprows=1,delimiter=",")

print(d1.dtype)
print(d1[:5,:])
```

```python
#loadtxt gives 2D array,
#genfromtxt gives 1D array so we dont use 2nd index
d2 = np.genfromtxt(filename, delimiter = ",",names = True,dtype=None)
print(d2.dtype)
print(d2[:5])
```

```python
#Pandas
d3 = pd.read_csv(filename)

print(d3.dtypes)

d3.head()
```
